```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame from the survey results
data = {
    "Timestamp": ["7/12/2024 10:01:23", "7/12/2024 10:15:47", "7/12/2024 10:30:05", "7/12/2024 10:45:12", 
                  "7/12/2024 11:00:34", "7/12/2024 11:15:56", "7/12/2024 11:31:07", "7/12/2024 11:45:45", 
                  "7/12/2024 12:00:23", "7/12/2024 12:15:47"],
    "Ease of Initial Profile Setup": [5, 4, 5, 5, 4, 5, 4, 4, 5, 4],
    "Understanding Recommendations": ["Yes", "Yes", "Yes", "Yes", "Yes", "Partially", "Yes", "Partially", "Yes", "Yes"],
    "Clarity of Portfolio Analysis": ["Good", "Good", "Excellent", "Good", "Good", "Good", "Good", "Good", "Excellent", "Good"],
    "Usefulness of Stock Price Forecasting": ["Very Useful", "Useful", "Extremely Useful", "Useful", "Useful", "Very Useful", "Extremely Useful", "Useful", "Very Useful", "Useful"],
    "Understanding Long-term Projected Returns": ["Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes", "Yes"],
    "Ease of Navigation": [5, 4, 5, 4, 4, 5, 4, 4, 5, 4],
    "Encountered Issues": ["No", "No", "No", "Yes", "No", "No", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data)

# Calculating averages and counts
ease_profile_setup_avg = df["Ease of Initial Profile Setup"].mean()
ease_navigation_avg = df["Ease of Navigation"].mean()

understanding_recommendations_count = df["Understanding Recommendations"].value_counts()
clarity_portfolio_analysis_count = df["Clarity of Portfolio Analysis"].value_counts()
usefulness_forecasting_count = df["Usefulness of Stock Price Forecasting"].value_counts()
understanding_returns_count = df["Understanding Long-term Projected Returns"].value_counts()
encountered_issues_count = df["Encountered Issues"].value_counts()

# Plotting charts
fig, axs = plt.subplots(3, 2, figsize=(14, 18))

# Chart for Ease of Initial Profile Setup
axs[0, 0].bar(df["Timestamp"], df["Ease of Initial Profile Setup"], color='skyblue')
axs[0, 0].set_title('Ease of Initial Profile Setup')
axs[0, 0].set_xlabel('Timestamp')
axs[0, 0].set_ylabel('Rating (1-5)')
axs[0, 0].set_xticklabels(df["Timestamp"], rotation=45, ha='right')

# Chart for Understanding Recommendations
axs[0, 1].bar(understanding_recommendations_count.index, understanding_recommendations_count.values, color='lightgreen')
axs[0, 1].set_title('Understanding Recommendations')
axs[0, 1].set_xlabel('Understanding')
axs[0, 1].set_ylabel('Count')

# Chart for Clarity of Portfolio Analysis
axs[1, 0].bar(clarity_portfolio_analysis_count.index, clarity_portfolio_analysis_count.values, color='coral')
axs[1, 0].set_title('Clarity of Portfolio Analysis')
axs[1, 0].set_xlabel('Clarity')
axs[1, 0].set_ylabel('Count')

# Chart for Usefulness of Stock Price Forecasting
axs[1, 1].bar(usefulness_forecasting_count.index, usefulness_forecasting_count.values, color='gold')
axs[1, 1].set_title('Usefulness of Stock Price Forecasting')
axs[1, 1].set_xlabel('Usefulness')
axs[1, 1].set_ylabel('Count')

# Chart for Understanding Long-term Projected Returns
axs[2, 0].bar(understanding_returns_count.index, understanding_returns_count.values, color='lightpink')
axs[2, 0].set_title('Understanding Long-term Projected Returns')
axs[2, 0].set_xlabel('Understanding')
axs[2, 0].set_ylabel('Count')

# Chart for Encountered Issues
axs[2, 1].bar(encountered_issues_count.index, encountered_issues_count.values, color='lightgray')
axs[2, 1].set_title('Encountered Issues')
axs[2, 1].set_xlabel('Encountered Issues')
axs[2, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Inputting investment preferences and risk tolerance": [5, 4, 5, 5, 4, 5, 4, 4, 5, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate average rating
average_rating = df.mean().values[0]

# Create a table
table_data = df.copy()
table_data.loc['Average'] = average_rating

# Create the chart
plt.figure(figsize=(10, 6))
plt.bar(df.index, df["Inputting investment preferences and risk tolerance"], color='blue')
plt.xlabel('Participants')
plt.ylabel('Rating')
plt.title('Inputting investment preferences and risk tolerance')
plt.xticks(df.index)
plt.ylim(0, 6)
plt.axhline(y=average_rating, color='r', linestyle='--', label=f'Average Rating: {average_rating:.2f}')
plt.legend()
plt.grid(True)
plt.show()

import ace_tools as tools; tools.display_dataframe_to_user(name="Inputting investment preferences and risk tolerance", dataframe=table_data)
```

```python
import matplotlib.pyplot as plt

# Data for the chart
responses = ['Extremely Useful', 'Very Useful', 'Useful']
counts = [2, 3, 5]

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(responses, counts, color=['green', 'blue', 'orange'])

# Annotate each bar with the count
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.1, int(yval), ha='center', va='bottom')

# Chart details
plt.title('Ease of Using the Stock Price Forecasting Tool')
plt.xlabel('Response')
plt.ylabel('Count')

plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

plt.tight_layout()
plt.show()
```
