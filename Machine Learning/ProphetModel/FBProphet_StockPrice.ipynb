```python
asset_list = ["^TNX", "^TYX", "AAPL", "ABBV", "ADBE", "AMZN", "AVGO", "BRK-B", "CRM","COST","CVX", "GOOG", "GOOGL", "HD", "JNJ", "JPM", "LLY", "MA", "META", "MRK", "MSFT", "NVDA","PG", "TSLA", "UNH", "V", "XOM","WELL", "O", "CCI"]
```

```python
from datetime import timedelta
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from prophet import Prophet

folder_path = "..\..\Data Collection & Processing\historical_prices_data" 
forecasts_folder = "..\..\Data Collection & Processing\stock_forecasts"
performance_folder = "..\..\Data Collection & Processing\stock_performance"
trend_folder = "..\..\Data Collection & Processing\stock_trends"

for asset in asset_list: 
  file_path = os.path.join(folder_path, f"{asset}.csv")
  # Import the top 25 sp500 stocks dataset into a Pandas Dataframe
  df = pd.read_csv(
    file_path,  
    index_col='Date', 
    infer_datetime_format=True, 
    parse_dates=True
  )
  df = df.reset_index()
  df = df.filter(['Date','Close'])
  df = df.dropna()
  # display (df)

  df.index[df['Close'] < 0]

  # # Plotting
  # plt.figure(figsize=(16, 8))
  # plt.plot(df['Date'], df['Close'], label='Close Price')
  # plt.title(f'Trend for {asset}', fontsize=20)
  # plt.xlabel('Date', fontsize=18)
  # plt.ylabel('Close Price USD ($)', fontsize=18)
  # plt.legend()
    
  # # Save the plot in the trend folder
  # plt.savefig(os.path.join(trend_folder, f"{asset}_trend.png"))
  # plt.close()  # Close the plot to free up memory

  # yt = yt-y(t-1) 
  df['Close Shift'] = df['Close'] - df['Close'].shift(1)
  y = df['Close Shift']

  # # Plotting
  # plt.figure(figsize=(16, 8))
  # plt.plot(df['Date'], df['Close Shift'], label='Daily Close Change')
  # plt.title(f'Stationary data for {asset}', fontsize=20)
  # plt.xlabel('Date', fontsize=18)
  # plt.ylabel('Change in Close Price USD ($)', fontsize=18)
  # plt.legend()

  # # Save the plot in the trend folder
  # plt.savefig(os.path.join(trend_folder, f"{asset}_stationary.png"))
  # plt.close()  # Close the plot to free up memory]

  # Label the columns ds and y so that the syntax is recognized by Prophet
  prophet_df = df.rename(columns = {'Date' : 'ds', 'Close Shift' : 'y'})

  # Sort the DataFrame by `ds` in ascending order
  prophet_df = prophet_df.sort_values('ds', ascending = True)

  # display(prophet_df)

  ## Split the Data
  train_data_len = int(0.8*len(df))
  # display(train_data_len)
  
  #80% for training and 20% for testing
  df_train = prophet_df[:train_data_len]
  df_test = prophet_df[train_data_len:]
  # display(df_train.head()), display(df_train.shape)

  model = Prophet(growth='linear')

  # Fit the Prophet model.
  model.fit(df_train)

  periods = len(df)-len(df_train)

  forecast = model.make_future_dataframe(periods= periods)

  # Make Prediction 
  prediction = model.predict(forecast)

  predictions = prediction.rename(columns={
    'yhat': 'Most Likely Case',
    'yhat_lower': 'Worst Case',
    'yhat_upper': 'Best Case'
  })

  # Ensure 'ds' column is of datetime type
  predictions['ds'] = pd.to_datetime(forecast['ds'])

  # Save to CSV without the index
  predictions.to_csv(os.path.join(forecasts_folder, f"{asset}_forecast.csv"), index=False)

  print(f"Forecast CSV for {asset} saved successfully.")


  ## prepare diagnostics report
  from prophet.diagnostics import cross_validation
  df_cv = cross_validation(model, initial='2013 days', period='1 days', horizon = '504 days')

  from prophet.diagnostics import performance_metrics
  df_p = performance_metrics(df_cv)
  
  df_p.to_csv(os.path.join(performance_folder, f"{asset}_performance.csv"), index=False)

  MSE = sum(df_p['mse'])/len(df_p)
  RMSE = sum(df_p['rmse'])/len(df_p)  
  display(f"{asset} - MSE: {MSE}, RMSE: {RMSE}")
```
