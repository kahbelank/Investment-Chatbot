### Aggressive Portfolio

```python
import os
import numpy as np
import pandas as pd
import yahoo_fin.stock_info as si
from datetime import datetime, timedelta
from MCPortfolioSimulator_LR import PortfolioSimulator
import matplotlib.pyplot as plt

# List of assets to fetch data for
asset_list = ['COST', 'LLY', 'META', 'MRK', 'NVDA', '^TNX', '^TYX', 'WELL', 'O', 'CCI']

# Define the end date for fetching the data
end_date = datetime.now()

# Function to fetch historical data and calculate returns
def fetch_data_and_calculate_returns(period_years):
    start_date = end_date - timedelta(days=period_years * 365)
    all_assets_df = pd.DataFrame()

    for asset in asset_list:
        print(f"Fetching data for {asset}...")
        df = si.get_data(asset, start_date=start_date, end_date=end_date, index_as_date=True)
        
        df.rename(columns={
            'open': 'open', 'high': 'high', 'low': 'low', 'close': 'close',
            'adjclose': 'adj close', 'volume': 'volume'
        }, inplace=True)
        
        if all_assets_df.empty:
            all_assets_df = df
            all_assets_df.columns = pd.MultiIndex.from_product([[asset], all_assets_df.columns])
        else:
            df.columns = pd.MultiIndex.from_product([[asset], df.columns])
            all_assets_df = pd.concat([all_assets_df, df], axis=1)
    
    all_assets_df.dropna(inplace=True)
    historical_returns = {}
    for asset in asset_list:
        close_prices = all_assets_df[asset]['close']
        daily_returns = close_prices.pct_change().dropna()
        avg_daily_return = daily_returns.mean()
        annual_return = ((1 + avg_daily_return) ** 252 - 1) * 100
        historical_returns[asset] = annual_return

    return all_assets_df, historical_returns

# Function to fetch benchmark data and calculate returns
def fetch_benchmark_data_and_calculate_returns(period_years, benchmark_symbol='^GSPC'):
    start_date = end_date - timedelta(days=period_years * 365)
    benchmark_df = si.get_data(benchmark_symbol, start_date=start_date, end_date=end_date, index_as_date=True)
    
    close_prices = benchmark_df['close']
    daily_returns = close_prices.pct_change().dropna()
    avg_daily_return = daily_returns.mean()
    annual_return = ((1 + avg_daily_return) ** 252 - 1) * 100
    annual_volatility = daily_returns.std() * np.sqrt(252)
    
    cumulative_return = (close_prices.iloc[-1] / close_prices.iloc[0]) - 1
    benchmark_stats = {
        'annual_return': annual_return,
        'annual_volatility': annual_volatility,
        'cumulative_return': cumulative_return,
        'cumulative_returns_series': (1 + daily_returns).cumprod()  # For plotting
    }
    
    return benchmark_stats, benchmark_df


def run_simulation_and_benchmark(period_years, initial_investment, weights, benchmark_symbol='^GSPC'):
    num_trading_days = period_years * 252
    all_assets_df, historical_returns = fetch_data_and_calculate_returns(period_years)
    benchmark_stats, benchmark_df = fetch_benchmark_data_and_calculate_returns(period_years, benchmark_symbol)

    sim_returns = PortfolioSimulator(
        portfolio_data=all_assets_df,
        annual_returns=historical_returns,
        weights=weights,
        num_simulation=500,
        num_trading_days=num_trading_days
    )

    cumulative_returns = sim_returns.calc_cumulative_return()
    sim_returns.plot_simulation()
    sim_returns.plot_distribution()

    summary = sim_returns.summarize_cumulative_return()
    print(summary)

    port_high_return = round(initial_investment * summary['95% CI Upper'], 2)
    port_low_return = round(initial_investment * summary['95% CI Lower'], 2)

    print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
          f" over the next {period_years} years will end within the range of"
          f" ${port_low_return} and ${port_high_return}")

    historical_cumulative_return = {}
    for asset in asset_list:
        close_prices = all_assets_df[asset]['close']
        cumulative_return = (close_prices.iloc[-1] / close_prices.iloc[0]) - 1
        historical_cumulative_return[asset] = cumulative_return

    historical_portfolio_return = sum(historical_cumulative_return[asset] * weight for asset, weight in zip(asset_list, weights))
    print(f"Historical Portfolio Cumulative Return for {period_years} years: {historical_portfolio_return * 100:.2f}%")

    simulated_mean_return = summary['mean']
    print(f"Simulated Mean Cumulative Return for {period_years} years: {simulated_mean_return * 100:.2f}%")

    historical_volatility = {}
    for asset in asset_list:
        close_prices = all_assets_df[asset]['close']
        daily_returns = close_prices.pct_change().dropna()
        annual_volatility = daily_returns.std() * np.sqrt(252)
        historical_volatility[asset] = annual_volatility

    historical_portfolio_volatility = np.sqrt(sum((weight * historical_volatility[asset]) ** 2 for asset, weight in zip(asset_list, weights)))
    print(f"Historical Portfolio Volatility for {period_years} years: {historical_portfolio_volatility * 100:.2f}%")

    simulated_volatility = summary['std']
    print(f"Simulated Portfolio Volatility for {period_years} years: {simulated_volatility * 100:.2f}%")

    print(f"Benchmark ({benchmark_symbol}) Cumulative Return for {period_years} years: {benchmark_stats['cumulative_return'] * 100:.2f}%")
    print(f"Benchmark ({benchmark_symbol}) Annual Return: {benchmark_stats['annual_return']:.2f}%")
    print(f"Benchmark ({benchmark_symbol}) Annual Volatility: {benchmark_stats['annual_volatility']:.2f}%")

    benchmark_cumulative_returns_aligned = benchmark_stats['cumulative_returns_series'].reindex(cumulative_returns.index).fillna(method='ffill')

    # Plot cumulative returns
    plt.figure(figsize=(14, 7))
    plt.plot(cumulative_returns.index, cumulative_returns.mean(axis=1) * initial_investment, label='Simulated Portfolio Mean')
    plt.plot(benchmark_cumulative_returns_aligned.index, benchmark_cumulative_returns_aligned * initial_investment, label=f'Benchmark ({benchmark_symbol})')
    plt.title(f'Cumulative Returns: Portfolio vs Benchmark over {period_years} years')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value ($)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Define weights
weights = [0.1, 0.2, 0.02, 0.2, 0.2, 0.08, 0.17, 0.01, 0.01, 0.01]

# Initial investment
initial_investment = 100000

# Run simulations and benchmarks for different periods
for period in [2, 5, 10, 15]:
    print(f"\n--- Simulating for {period} years ---")
    run_simulation_and_benchmark(period, initial_investment, weights)
```
